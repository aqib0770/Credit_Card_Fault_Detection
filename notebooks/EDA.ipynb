```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('creditCardFraud_28011964_120214.csv')
df.head()
```

```python
df['default payment next month'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X=df.drop('default payment next month',axis=1)
y=df['default payment next month']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
reg=LogisticRegression()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE(random_state=42)
X_smote,y_smote=smote.fit_resample(X,y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_smote,y_smote,test_size=0.3,random_state=42)
```

```python
reg2=LogisticRegression()
reg2.fit(X_train,y_train)
```

```python
y_pred2=reg2.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "SVM": SVC(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier()
}
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"{model_name}:\n", classification_report(y_test, y_pred))
    print(accuracy_score(y_test,y_pred),'\n')
    print('________________________________________')
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'n_estimators':[100,200,300,400,500],
        'criterion':['gini','entropy'],
        'max_depth':[3,5,7,9],
        'max_features':['auto','sqrt','log2']}
```

```python
gcv=GridSearchCV(rfc,params,verbose=3)
```

```python
gcv.fit(X_train,y_train)
y_pred=gcv.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
gcv.best_params_
```

```python
rfc2=RandomForestClassifier(criterion='gini',max_depth=9,max_features='log2',n_estimators=200)
rfc2.fit(X_train,y_train)
y_pred=rfc2.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

- With above observation, I found that RandomForestClassifier is the best model for this problem statement.
- With GridSearchCV, I found that the best parameters for this model are:
<br>{'criterion': 'gini',
 'max_depth': 9,
 'max_features': 'log2',
